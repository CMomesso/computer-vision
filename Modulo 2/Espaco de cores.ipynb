```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
%matplotlib inline
```

```python
img = cv2.imread("data/camis.jpg")
```

```python
plt.imshow(img)
```

```python
img2 = img[..., ::-1]
```

```python
plt.imshow(img2)
```

```python
r, g, b = cv2.split(img2)

plt.figure(figsize=[20,5])

plt.subplot(141);
plt.imshow(r);
plt.title("Vermelho")

plt.subplot(142);
plt.imshow(g);
plt.title("Verde");

plt.subplot(143);
plt.imshow(b);
plt.title("Azul")

plt.subplot(144);
img_merged = cv2.merge((r,g, b))
plt.imshow(img_merged)
```
